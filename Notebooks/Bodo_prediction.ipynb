```python
import pandas as pd
import pickle

# Load training data with text features
train_df = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/Bodo_Prediction/Bodo_train_data_with_text_features.csv')

# Load test data with OCR features
test_df = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/Bodo_Prediction/Bodo_test_data_with_ocr_features.csv')

# Load training image features
with open('/content/drive/MyDrive/Colab Notebooks/Bodo_Prediction/Bodo_image_features.pkl', 'rb') as f:
    train_image_features = pickle.load(f)

# Load test image features
with open('/content/drive/MyDrive/Colab Notebooks/Bodo_Prediction/Bodo_test_image_features.pkl', 'rb') as f:
    test_image_features = pickle.load(f)

print("Data loading complete.")
```

```python
# Convert image features numpy arrays to DataFrames
train_image_features_df = pd.DataFrame(train_image_features)
test_image_features_df = pd.DataFrame(test_image_features)

# Add a simple index column to both image feature DataFrames and the text DataFrames for merging
train_image_features_df['merge_index'] = train_image_features_df.index
test_image_features_df['merge_index'] = test_image_features_df.index
train_df['merge_index'] = train_df.index
test_df['merge_index'] = test_df.index


# Combine text and image features for training data
train_combined_df = pd.merge(train_df, train_image_features_df, on='merge_index').drop('merge_index', axis=1)

# Combine text and image features for test data
test_combined_df = pd.merge(test_df, test_image_features_df, on='merge_index').drop('merge_index', axis=1)

print("Feature combining complete.")
display(train_combined_df.head())
display(test_combined_df.head())
```

```python
# Assuming the labels are in the 'Sentiment', 'Sarcasm', 'Vulgar', 'Abuse', and 'Target' columns of the training data
label_columns = ['Sentiment', 'Sarcasm', 'Vulgar', 'Abuse', 'Target']
train_labels = train_combined_df[label_columns]

print("Training labels loaded.")
display(train_labels.head())
```

```python
from imblearn.over_sampling import SMOTE
import numpy as np

# Prepare training features by dropping non-feature columns
# Ensure consistency with columns used for model training later
X_train = train_combined_df.drop(['Ids', 'OCR', 'text_features', 'Sentiment', 'Sarcasm', 'Vulgar', 'Abuse', 'Target'], axis=1)


# Apply SMOTE to each label separately
balanced_train_data = {}
for label_col in label_columns:
    print(f"Applying SMOTE for {label_col}...")
    y_train = train_labels[label_col]

    # Handle potential missing values in labels (e.g., for 'Target') before applying SMOTE
    if y_train.isnull().any():
        print(f"Warning: Missing values found in {label_col}. Dropping rows with missing labels for SMOTE.")
        X_train_subset = X_train[y_train.notnull()]
        y_train_subset = y_train[y_train.notnull()]
    else:
        X_train_subset = X_train
        y_train_subset = y_train

    # Check if there's more than one class after handling missing values
    if len(y_train_subset.unique()) > 1:
        # Determine the number of samples in the minority class
        value_counts = y_train_subset.value_counts()
        minority_class_size = value_counts.min()

        # Adjust n_neighbors for SMOTE
        # n_neighbors must be less than or equal to the number of samples in the minority class
        # If minority class size is 1, SMOTE cannot be applied, so we skip it.
        # If minority class size is > 1, n_neighbors can be up to minority_class_size - 1, minimum 1.
        n_neighbors = max(1, minority_class_size - 1)

        if n_neighbors >= 1 and minority_class_size > 1:
            smote = SMOTE(random_state=42, k_neighbors=n_neighbors)
            X_resampled, y_resampled = smote.fit_resample(X_train_subset, y_train_subset)
            balanced_train_data[label_col] = (X_resampled, y_resampled)
            print(f"  Original shape: {X_train_subset.shape}, Resampled shape: {X_resampled.shape}")
            print(f"  Resampled value counts:\n{y_resampled.value_counts()}")
        else:
            print(f"  Skipping SMOTE for {label_col} due to too few samples in minority class ({minority_class_size}).")
            balanced_train_data[label_col] = (X_train_subset, y_train_subset) # Keep original data if SMOTE is skipped
    else:
        print(f"  Skipping SMOTE for {label_col} as only one class is present.")
        balanced_train_data[label_col] = (X_train_subset, y_train_subset) # Keep original data if only one class

print("SMOTE application complete.")
```

```python
import xgboost as xgb
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
from sklearn.metrics import f1_score # Import F1-score

# Train a separate model for each label using balanced data
models = {}
for label_col in label_columns:
    print(f"Training model for {label_col}...")

    # Get balanced data for the current label
    if label_col in balanced_train_data:
      X_train_balanced, y_train_balanced = balanced_train_data[label_col]
    else:
      # If SMOTE was skipped, use the original subset
      y_train = train_labels[label_col]
      if y_train.isnull().any():
          X_train_balanced = X_train[y_train.notnull()]
          y_train_balanced = y_train[y_train.notnull()]
      else:
          X_train_balanced = X_train
          y_train_balanced = y_train


    # Encode labels if they are not numerical
    if y_train_balanced.dtype == 'object' or y_train_balanced.dtype == 'category':
        label_encoder = LabelEncoder()
        y_train_encoded = label_encoder.fit_transform(y_train_balanced)
        models[f'{label_col}_label_encoder'] = label_encoder # Store the encoder
    else:
        y_train_encoded = y_train_balanced

    # Initialize and train XGBoost classifier
    # Using 'multi:softmax' objective and number of classes
    # eval_metric can be changed to f1_score or 'auc' depending on preference and binary/multiclass
    # For multi-class with potential imbalance, 'mlogloss' is reasonable, but F1-score is better for evaluation
    model = xgb.XGBClassifier(objective='multi:softmax', num_class=len(set(y_train_encoded)), eval_metric='mlogloss', use_label_encoder=False)
    model.fit(X_train_balanced, y_train_encoded)
    models[label_col] = model # Store the trained model

print("Model training complete.")
```

```python
# Prepare test data for prediction
X_test = test_combined_df.drop(['Ids', 'OCR', 'tokenized_ocr', 'ocr_features'], axis=1) # Assuming these columns are not features

# Align test set columns with training set columns
train_cols = X_train.columns # Use original X_train columns for alignment
test_cols = X_test.columns

missing_in_test = set(train_cols) - set(test_cols)
for c in missing_in_test:
    X_test[c] = 0  # Add missing columns to test set with default value 0

missing_in_train = set(test_cols) - set(train_cols)
# Drop columns in test set that are not in training set
X_test = X_test.drop(columns=missing_in_train)


X_test = X_test[train_cols] # Ensure the order of columns is the same

# Make predictions for each label
test_predictions = pd.DataFrame()
test_predictions['Ids'] = test_combined_df['Ids'] # Keep track of Ids

for label_col in label_columns:
    print(f"Making predictions for {label_col}...")
    model = models[label_col]

    # If the original training labels were encoded, use the stored encoder to inverse transform predictions
    if f'{label_col}_label_encoder' in models:
        label_encoder = models[f'{label_col}_label_encoder']
        # Need to handle potential missing values in test data if they were handled in training
        # For simplicity here, we assume no missing values in test features
        predictions_encoded = model.predict(X_test)
        predictions = label_encoder.inverse_transform(predictions_encoded)
    else:
        predictions = model.predict(X_test)

    test_predictions[label_col] = predictions

print("Predictions complete.")
display(test_predictions.head())
```

```python
# Save the predictions DataFrame to a new CSV
test_predictions.to_csv('/content/drive/MyDrive/Colab Notebooks/Bodo_Prediction/Bodo_predictions.csv', index=False)
print("Predictions saved to a new CSV.")
```
